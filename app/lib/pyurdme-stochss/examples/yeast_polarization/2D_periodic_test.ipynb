```python
import pyurdme
from matplotlib import cm as CM
from IPython.display import display, clear_output
import time
```

```python
num_vox = 30
class testDiffusion2D(pyurdme.URDMEModel):
    def __init__(self):
        pyurdme.URDMEModel.__init__(self,"test1D")
        X = pyurdme.Species(name="X",  diffusion_constant=0.001)
        self.add_species([X])
        self.mesh = pyurdme.URDMEMesh.generate_unit_square_mesh(nx=num_vox, ny=num_vox, periodic=True)
        self.set_initial_condition_place_near({X:num_vox*num_vox*10}, (0.1,0.1))
        #self.set_initial_condition_scatter({X:1000})
        #self.set_initial_condition_distribute_uniformly({X:1000})
        self.timespan(range(101))
```

```python
model = testDiffusion2D()
result = model.run()
```

```python
x_vals = model.mesh.coordinates()[:, 0]
y_vals = model.mesh.coordinates()[:, 1]
z_vals = result.get_species("X")
```
