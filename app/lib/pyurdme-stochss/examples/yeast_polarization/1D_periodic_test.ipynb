```python
import pyurdme
```

```python
class testDiffusion1D(pyurdme.URDMEModel):
    def __init__(self):
        pyurdme.URDMEModel.__init__(self,"test1D")
        X = pyurdme.Species(name="X",  diffusion_constant=0.001)
        self.add_species([X])
        self.mesh = pyurdme.URDMEMesh.generate_unit_interval_mesh(nx=100, periodic=False)
        self.set_initial_condition_place_near({X:1000}, 0.1)
        #self.set_initial_condition_scatter({X:1000})
        #self.set_initial_condition_distribute_uniformly({X:1000})
        self.timespan(range(101))
```

```python
model = testDiffusion1D()
result = model.run()
x_vals = model.mesh.coordinates()[:, 0]
z_vals = result.get_species("X")
```

```python
plt.imshow(z_vals)
plt.clim(0,20)
plt.xlabel("space")
plt.ylabel("time")
plt.colorbar()
```
